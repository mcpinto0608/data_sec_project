# Isolation Forest â€“ First Tuning Phase

This notebook performs the first parameter tuning of the Isolation Forest model trained on clean benign traffic. It explores combinations of key hyperparameters and evaluates each configuration using F1-score. The selected model is then analyzed across multiple thresholds to find the optimal trade-off between attack detection and false positives. The result is a tuned baseline model suitable for further refinement or evaluation on new data.


# 1. Imports and Data Preparation

Import all necessary libraries for model training, evaluation, and visualization. Load the scaled dataset and labels that were prepared during the initial training phase. These will be used to train and evaluate multiple Isolation Forest configurations.


```python
import pandas as pd
import numpy as np
import joblib
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest
from sklearn.metrics import precision_score, recall_score, f1_score, classification_report, confusion_matrix
from sklearn.decomposition import PCA
import os
from datetime import datetime
```

```python
#Load Data
data = pd.read_csv("../../data/cleaned_cic.csv")
data.columns = data.columns.str.strip().str.lower()
scaler = joblib.load("../../models/first_model/scaler.joblib")

X = data.drop("attack", axis=1)
y = data["attack"]
X_scaled = scaler.transform(X)
```

# 2. Model Training Loop and Scoring Function

Define a scoring function that trains an Isolation Forest model with a given parameter combination. Evaluate each configuration by computing predictions and calculating F1-score using the default threshold. This loop tests all candidate settings in the defined parameter grid.


```python
# Define the function to loop through the candidates and tweak the parameters
def evaluate_iforest(model, X_scaled, y, verbose=True):
    scores = model.decision_function(X_scaled)
    thresholds = np.linspace(min(scores), max(scores), 50)
    results = []

    for t in thresholds:
        preds = [1 if s < t else 0 for s in scores]
        p = precision_score(y, preds, zero_division=0)
        r = recall_score(y, preds, zero_division=0)
        f1 = f1_score(y, preds, zero_division=0)
        results.append((t, p, r, f1))

    df = pd.DataFrame(results, columns=["Threshold", "Precision", "Recall", "F1"])
    best = df.loc[df["F1"].idxmax()]
    
    if verbose:
        print("ðŸ” Best Threshold: {:.4f}".format(best["Threshold"]))
        print("Precision: {:.2f}, Recall: {:.2f}, F1: {:.2f}".format(best["Precision"], best["Recall"], best["F1"]))

    return best["F1"], best["Threshold"], model
```

```python
candidates = [
    {"n_estimators": 100, "max_samples": 0.8, "contamination": 0.01},
    {"n_estimators": 200, "max_samples": 0.8, "contamination": 0.015},
    {"n_estimators": 300, "max_samples": 1.0, "contamination": 0.02},
    {"n_estimators": 300, "max_samples": 0.6, "contamination": 0.025},
    {"n_estimators": 500, "max_samples": 0.5, "contamination": 0.03},
    {"n_estimators": 150, "max_samples": 1.0, "contamination": 0.015},
    {"n_estimators": 200, "max_samples": 0.9, "contamination": 0.02},
    {"n_estimators": 400, "max_samples": 0.7, "contamination": 0.015},
    {"n_estimators": 100, "max_samples": 0.9, "contamination": 0.025},
    {"n_estimators": 250, "max_samples": 0.6, "contamination": 0.02},
    {"n_estimators": 350, "max_samples": 1.0, "contamination": 0.01},
    {"n_estimators": 450, "max_samples": 0.8, "contamination": 0.015},
]
```

```python
best_f1 = 0
best_model = None
best_params = None
best_threshold = None
model_results = []

#Loop through the candidates
for i, params in enumerate(candidates):
    print("="*40)
    print(f" Training model #{i+1}")
    print(" Params:", params)

    model = IsolationForest(**params, random_state=42)
    model.fit(X_scaled[y == 0])

    print(" Evaluating model...")
    f1, threshold, _ = evaluate_iforest(model, X_scaled, y)
    print(f" F1 Score: {f1:.4f} | Threshold: {threshold:.4f}")

    model_results.append({
        "params": params,
        "f1": f1,
        "threshold": threshold
    })

    #Save best model so far
    if f1 > best_f1:
        print(" New best model found!")
        best_f1 = f1
        best_model = model
        best_params = params
        best_threshold = threshold
```

# 3. Select Best Model Based on F1 Score

Aggregate the results from all tested configurations into a DataFrame. Identify and retain the model that achieved the highest F1-score. This model will be used for threshold tuning and final evaluation.


```python
# Collect threshold scores
results = []
thresholds = np.linspace(min(scores), max(scores), 50)

for t in thresholds:
    preds = [1 if s < t else 0 for s in scores]
    precision = precision_score(y, preds, zero_division=0)
    recall = recall_score(y, preds, zero_division=0)
    f1 = f1_score(y, preds, zero_division=0)
    results.append((t, precision, recall, f1))

# Create DataFrame
results_df = pd.DataFrame(results, columns=["Threshold", "Precision", "Recall", "F1"])

# Find best by F1
best_row = results_df.loc[results_df["F1"].idxmax()]
```

# 4. Visualization and Model Saving

Plot these metrics and identify the best threshold. Save the final model, threshold, and performance summary for downstream use.


```python
# Plot
plt.figure(figsize=(12, 5))
plt.plot(results_df["Threshold"], results_df["Precision"], label="Precision", color="gold")
plt.plot(results_df["Threshold"], results_df["Recall"], label="Recall", color="darkorange")
plt.plot(results_df["Threshold"], results_df["F1"], label="F1 Score", color="crimson")
plt.axvline(x=best_row["Threshold"], color="blue", linestyle="--", label=f'Best Threshold ({best_row["Threshold"]:.3f})')

plt.xlabel("Threshold")
plt.ylabel("Score")
plt.title("ðŸ“Š Threshold Tuning Results for Best Model")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Final predictions and confusion matrix
final_preds = (scores < best_threshold).astype(int)
conf_matrix = confusion_matrix(y, final_preds)

# Print confusion matrix
print("\nFinal Confusion Matrix:")
print(conf_matrix)

# Classification report table
print("\nðŸ§¾ Classification Report:\n")
print(classification_report(y, final_preds, target_names=["Benign", "Attack"]))


# Breakdown
tn, fp = conf_matrix[0]
fn, tp = conf_matrix[1]
print("\nDetailed Results:")
print(f"True Negatives (Benign correct):      {tn:,}")
print(f"False Positives (Benign as attack):   {fp:,}")
print(f"False Negatives (Missed attacks):     {fn:,}")
print(f"True Positives (Attacks detected):    {tp:,}")
```

```python
import joblib
from pathlib import Path
import pandas as pd

# Create save folder (if not exists)
tuning_dir = Path("../../models/first_tuning")
tuning_dir.mkdir(parents=True, exist_ok=True)

# Save model (always overwrite)
model_path = tuning_dir / "iforest_tuned.joblib"
joblib.dump(best_model, model_path)

# Save CSV with tuning summary (overwrite)
summary = {
    "params": best_params,
    "threshold": best_threshold,
    "f1_score": best_f1
}
csv_path = tuning_dir / "tuning_summary.csv"
pd.DataFrame([summary]).to_csv(csv_path, index=False)

# Confirm
print(f"\nâœ… Best model saved to:     {model_path}")
print(f"ðŸ“ Tuning summary saved to: {csv_path}")
print("ðŸ“ˆ F1 Score:", best_f1)
print("ðŸ“ Threshold:", best_threshold)
print("ðŸ§ª Params:", best_params)
```
