```python
# Second notebook of model tuning:
# Load the clean dataset
# Use permutation importance to extract the top 15 most useful features for model training
# Use the best model to date to train on the top features and save the new model
# Result: 
#Improved model and metadata saved:
#../models/iforest_retrained_top15_f1_0.647_20250402_145446.joblib
#../models/scaler_top15_20250402_145446.joblib
#../models/params_top15_20250402_145446.json
#../models/extra_training_log.csv
```

```python
import pandas as pd
import numpy as np
import joblib
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest
from sklearn.metrics import precision_score, recall_score, f1_score, classification_report, confusion_matrix
from sklearn.inspection import permutation_importance
from datetime import datetime
```

```python
# Load dataset
data = pd.read_csv("../data/clean/cleaned_cic.csv")
data.columns = data.columns.str.strip().str.lower()

X = data.drop("attack", axis=1)
y = data["attack"]

# Load scaler and model
scaler = joblib.load("../models/scaler.joblib")
model = joblib.load("../models/iforest_tuned_f1_0.554_20250401_153548.joblib")  # Use your tuned model file name here

X_scaled = scaler.transform(X)
```

```python
print(" Sampling 250,000 rows for permutation importance...")

# Sample 250,000 rows for better performance with accuracy
sample_idx = np.random.choice(len(X_scaled), size=250000, replace=False)
X_sample = X_scaled[sample_idx]
y_sample = y.iloc[sample_idx]

# Compute permutation importances
results = permutation_importance(
    model, X_sample, y_sample,
    scoring="f1_macro",
    n_repeats=5,
    random_state=42,
    n_jobs=-1
)

# Create and sort DataFrame
importances_df = pd.DataFrame({
    "feature": X.columns,
    "importance": results.importances_mean
}).sort_values("importance", ascending=False)

# Plot top 15
importances_df.head(15).plot.barh(x='feature', y='importance', figsize=(10, 6))
plt.gca().invert_yaxis()
plt.title("‚ú® Top 15 Important Features (Sampled)")
plt.xlabel("Mean Importance (F1 Macro)")
plt.ylabel("Feature")
plt.tight_layout()
plt.show()
```

```python
# Select top N features
top_n = 15
top_features = importances_df.head(top_n)["feature"].tolist()

print(f"\n‚ú® Retraining model with top {top_n} features...")

# Subset the data
X_top = X[top_features]
X_top_scaled = scaler.fit_transform(X_top)

# Retrain using best hyperparameters
params = model.get_params()
params.pop("random_state", None)  # Remove to avoid duplication
new_model = IsolationForest(**params, random_state=42)

print(" Fitting model on top features (this may take a bit)...")
new_model.fit(X_top_scaled[y == 0])
print(" Model retrained successfully!")
```

```python
print(" Re-tuning detection threshold...")

scores = new_model.decision_function(X_top_scaled)
thresholds = np.linspace(min(scores), max(scores), 50)

tuning_results = []
for t in thresholds:
    preds = [1 if s < t else 0 for s in scores]
    f1 = f1_score(y, preds, zero_division=0)
    tuning_results.append((t, f1))

tuning_df = pd.DataFrame(tuning_results, columns=["Threshold", "F1"])
best_row = tuning_df.loc[tuning_df["F1"].idxmax()]

best_threshold = best_row["Threshold"]
best_f1 = best_row["F1"]

print(f" New Best Threshold: {best_threshold:.4f} | F1: {best_f1:.4f}")
```

```python
preds = [1 if s < best_threshold else 0 for s in scores]

print("üìä Final Evaluation Report:")
print(classification_report(y, preds, target_names=["Benign", "Attack"]))

print("üß© Final Confusion Matrix:")
print(confusion_matrix(y, preds))
```

```python
from pathlib import Path
import json

# Create timestamped filenames
timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
model_filename = f"../models/iforest_retrained_top{top_n}_f1_{best_f1:.3f}_{timestamp}.joblib"
scaler_filename = f"../models/scaler_top{top_n}_{timestamp}.joblib"
params_filename = f"../models/params_top{top_n}_{timestamp}.json"
log_filename = "../models/extra_training_log.csv"

# Save model, scaler and params
joblib.dump(new_model, model_filename)
joblib.dump(scaler, scaler_filename)
with open(params_filename, "w") as f:
    json.dump(params, f, indent=4)

# Prepare log metadata
log_entry = pd.DataFrame([{
    "timestamp": timestamp,
    "f1_score": best_f1,
    "threshold": best_threshold,
    "top_features": ", ".join(top_features),
    "n_features": len(top_features),
    "model_filename": model_filename,
    "scaler_filename": scaler_filename,
    "params_file": params_filename
}])

# Append to existing CSV log
try:
    log_df = pd.read_csv(log_filename)
    log_df = pd.concat([log_df, log_entry], ignore_index=True)
except FileNotFoundError:
    log_df = log_entry

log_df.to_csv(log_filename, index=False)

# Confirm to user
print(f" Improved model and metadata saved:\n {model_filename}\n {scaler_filename}\n {params_filename}\n {log_filename}")
```
