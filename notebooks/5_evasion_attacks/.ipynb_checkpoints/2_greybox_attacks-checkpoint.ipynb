# Grey-box Evasion Attacks on Random Forest IDS Model

In this notebook, we simulate adversarial evasion attacks under **grey-box assumptions**:

- The attacker knows the features used in detection
- They can query the model to observe predictions (labels or probabilities)
- But they have no access to the modelâ€™s internal structure or learned feature weights

We'll craft multiple attacks and measure their effectiveness using:
- Accuracy, precision, recall, F1-score
- Evasion rate (% of attacks misclassified as benign)
- Confusion matrix and classification report


```python
import os
import pandas as pd
import numpy as np
import joblib
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
from sklearn.metrics import (
    classification_report, confusion_matrix, accuracy_score,
    precision_score, recall_score, f1_score, roc_auc_score
)

# Suppress warnings
warnings.simplefilter("ignore", category=FutureWarning)
```

```python
# === Paths
model_path = '../../models/supervised_real_traffic_test/model.joblib'
benign_path = '../../real_data/benign'
attack_path = '../../real_data/attack'
reference_csv = '../../data/cleaned_cic.csv'

# === Load model
model = joblib.load(model_path)

# === Load function
def load_labeled_csvs(folder, label):
    dfs = []
    for fname in os.listdir(folder):
        if fname.endswith('.csv'):
            df = pd.read_csv(os.path.join(folder, fname))
            df['attack'] = label
            dfs.append(df)
    return pd.concat(dfs, ignore_index=True) if dfs else pd.DataFrame()

# === Load traffic
benign_df = load_labeled_csvs(benign_path, 0)
attack_df = load_labeled_csvs(attack_path, 1)
df_all = pd.concat([benign_df, attack_df], ignore_index=True)

print(f"Loaded: {len(benign_df)} benign, {len(attack_df)} attack samples")
```

```python
# === Align to training features
ref_df = pd.read_csv(reference_csv)
expected_features = ref_df.select_dtypes(include=[np.number]).drop(columns=['attack']).columns.tolist()

X = df_all[expected_features].fillna(0).astype(float)
y = df_all['attack'].astype(int)

print(f"Features aligned. Input shape: {X.shape}")
```

```python
# === Predict on original data ===
y_pred = model.predict(X)
y_prob = model.predict_proba(X)[:, 1]

# === Metrics before adversarial modification ===
acc = accuracy_score(y, y_pred)
prec = precision_score(y, y_pred)
rec = recall_score(y, y_pred)
f1 = f1_score(y, y_pred)
auc = roc_auc_score(y, y_prob)

print("Pre-Attack Metrics:")
print(f"Accuracy:  {acc:.4f}")
print(f"Precision: {prec:.4f}")
print(f"Recall:    {rec:.4f}")
print(f"F1-Score:  {f1:.4f}")
print(f"ROC AUC:   {auc:.4f}")

# === Metrics and confusion matrix
print("Classification Report â€“ After Attack 1:")
print(classification_report(y, y_pred, target_names=["Benign", "Attack"]))

# === Confusion Matrix ===
cm = confusion_matrix(y, y_pred)
plt.figure(figsize=(6,5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Benign', 'Attack'], yticklabels=['Benign', 'Attack'])
plt.xlabel("Predicted")
plt.ylabel("True")
plt.title("Confusion Matrix â€“ Before Attack")
plt.tight_layout()
plt.show()
```

```python
def evaluate_attack(y_true, y_pred, y_scores, title=""):
    acc = accuracy_score(y_true, y_pred)
    prec = precision_score(y_true, y_pred)
    rec = recall_score(y_true, y_pred)
    f1 = f1_score(y_true, y_pred)
    auc = roc_auc_score(y_true, y_scores)

    attack_mask = (y_true == 1)
    evaded = np.sum((y_true == 1) & (y_pred == 0))
    total_attacks = np.sum(attack_mask)
    evasion_rate = evaded / total_attacks if total_attacks > 0 else 0

    print(f"Classification Report â€“ {title}")
    print(classification_report(y_true, y_pred, target_names=["Benign", "Attack"]))
    print(f"Accuracy:         {acc:.4f}")
    print(f"Precision:        {prec:.4f}")
    print(f"Recall:           {rec:.4f}")
    print(f"F1-Score:         {f1:.4f}")
    print(f"ROC AUC:          {auc:.4f}")
    print(f"Evasion Rate:  {evasion_rate:.2%} ({evaded} of {total_attacks} attacks)")

    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(6,5))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Benign', 'Attack'], yticklabels=['Benign', 'Attack'])
    plt.xlabel("Predicted")
    plt.ylabel("True")
    plt.title(f"Confusion Matrix â€“ {title}")
    plt.tight_layout()
    plt.show()
```

## Grey-box Attack 1: Query-Based Boundary Drift

The attacker:
- Knows the features used in detection
- Can observe whether inputs are labeled "benign" or "attack"
- Knows that benign traffic is typically centered around certain value ranges

Strategy:
- Slightly nudge the attack samples **closer to the mean** of benign samples
- Try to "blend in" just enough to cross the detection boundary
- This is a cautious, exploratory attacker


```python
# === Clone inputs
X_adv1 = X.copy()
y_adv1 = y.copy()

attack_indices = y_adv1[y_adv1 == 1].index
benign_indices = y_adv1[y_adv1 == 0].index

# === Use only top 5 features
importances = pd.Series(model.feature_importances_, index=X.columns)
top5_feats = importances.sort_values(ascending=False).head(5).index.tolist()
benign_means = X_adv1.loc[benign_indices, top5_feats].mean()

# === Light blend: 98% original + 2% benign mean (on top 5 features only)
for feat in top5_feats:
    X_adv1.loc[attack_indices, feat] = (
        0.98 * X_adv1.loc[attack_indices, feat] +
        0.02 * benign_means[feat]
    )

print(f"Drifted {len(top5_feats)} features lightly across {len(attack_indices)} attack samples")


y_pred_adv1 = model.predict(X_adv1)
y_prob_adv1 = model.predict_proba(X_adv1)[:, 1]
evaluate_attack(y, y_pred_adv1, y_prob_adv1, title="Grey-box Attack 1 â€“ Light Drift Toward Benign")
```

## Grey-box Attack 2: Benign Cluster Imitation (Top Features)

We now simulate a more informed attacker who:
- Has access to prior benign traffic
- Understands which features are likely used
- Modifies only the most influential features to stay inside "safe" benign zones

This attack is still grey-box: it doesnâ€™t require internal model access.


```python
# === Clone
X_adv2 = X.copy()
y_adv2 = y.copy()
attack_indices = y_adv2[y_adv2 == 1].index
benign_indices = y_adv2[y_adv2 == 0].index

# === Top 4 features
top_feats = importances.sort_values(ascending=False).head(4).index.tolist()
top3, top1 = top_feats[:3], top_feats[3:]

# === Sample benign values
benign_samples_3 = X_adv2.loc[benign_indices, top3].sample(
    n=len(attack_indices), replace=True, random_state=42
).reset_index(drop=True)
benign_samples_1 = X_adv2.loc[benign_indices, top1].sample(
    n=len(attack_indices), replace=True, random_state=42
).reset_index(drop=True)

# === Replace top 3
X_adv2.loc[attack_indices, top3] = benign_samples_3.values

# === Blend top 4th feature: 90% original + 10% benign
for feat in top1:
    original = X_adv2.loc[attack_indices, feat]
    benign = benign_samples_1[feat]
    X_adv2.loc[attack_indices, feat] = 0.9 * original + 0.1 * benign

print(f"Hybrid attack: top 3 replaced, top 4th blended 90/10 across {len(attack_indices)} attack samples")


y_pred_adv2 = model.predict(X_adv2)
y_prob_adv2 = model.predict_proba(X_adv2)[:, 1]
evaluate_attack(y, y_pred_adv2, y_prob_adv2, title="Grey-box Attack 2 â€“ Benign Cluster Imitation")
```

## Grey-box Attack 3 â€“ Surrogate Model Guided Evasion

This attack simulates an adversary who:
- Has access to real network traffic samples labeled as benign or attack
- Trains their own surrogate detection model (e.g., Random Forest)
- Uses that model to generate samples that evade detection

They then submit these samples to the real model, assuming some level of transferability.



```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.utils import shuffle

# Sample + inject label noise
benign = X[y == 0].sample(n=100_000, random_state=42)
attacks = X[y == 1].sample(n=20_000, random_state=42)
noisy_y_attacks = pd.Series([1]*20_000)
flip_mask = noisy_y_attacks.sample(frac=0.4, random_state=42).index
noisy_y_attacks.loc[flip_mask] = 0

# Combine and shuffle
X_surrogate = pd.concat([benign, attacks])
y_surrogate = pd.concat([
    pd.Series([0]*len(benign), index=benign.index),
    noisy_y_attacks.set_axis(attacks.index)
])
X_surrogate, y_surrogate = shuffle(X_surrogate, y_surrogate, random_state=42)

# Train
surrogate_rf = RandomForestClassifier(
    n_estimators=30,
    max_depth=3,
    max_features=0.4,
    random_state=42,
    n_jobs=-1
)
surrogate_rf.fit(X_surrogate, y_surrogate)
print(f"Trained weak surrogate on {len(X_surrogate)} samples")
```

```python
# Predict surrogate probabilities
surrogate_probs = surrogate_rf.predict_proba(X)[:, 1]
evading_indices = (y == 1) & (surrogate_probs < 0.5)
evading_pool = X.loc[evading_indices].copy()
print(f"Collected {len(evading_pool)} surrogate-evading attack samples")
```

```python
# Clone and prepare attack rows
X_adv3 = X.copy()
y_adv3 = y.copy()
attack_indices = y_adv3[y_adv3 == 1].index

# Top features by surrogate importance
importances = pd.Series(surrogate_rf.feature_importances_, index=X.columns)
top5 = importances.sort_values(ascending=False).head(5).index.tolist()
blend2 = importances.sort_values(ascending=False).iloc[5:7].index.tolist()
print(f"Substituting: {top5}, blending: {blend2}")

# Sample from evasion pool
sub_values = evading_pool[top5].sample(n=len(attack_indices), replace=True, random_state=42).reset_index(drop=True)
blend_values = evading_pool[blend2].sample(n=len(attack_indices), replace=True, random_state=42).reset_index(drop=True)

# Substitute
X_adv3.loc[attack_indices, top5] = sub_values.values

# Blend additional features (90% original, 10% evading)
for feat in blend2:
    X_adv3.loc[attack_indices, feat] = (
        0.9 * X_adv3.loc[attack_indices, feat].values +
        0.1 * blend_values[feat].values
    )
```

```python
# Build final test set
X_adv3_final = pd.concat([
    X[y == 0],                    # Benign
    X_adv3.loc[attack_indices]   # Modified attacks
])
y_adv3_final = pd.concat([
    y[y == 0],
    y.loc[attack_indices]
])

# Evaluate
y_pred = model.predict(X_adv3_final)
y_prob = model.predict_proba(X_adv3_final)[:, 1]

evaluate_attack(
    y_adv3_final, y_pred, y_prob,
    title="Grey-box Attack 3 â€“ Final Substitution + Mimicry"
)
```

## Grey-box Attack 5 â€“ Class Blending (Benign + Attack)

This attack simulates a grey-box adversary who:
- Observes both benign and attack traffic patterns
- Blends them together to create synthetic evasive flows
- Leverages benign structure to fool the model, while preserving enough of the attack signal


```python
# Clone labels
X_adv5 = X.copy()
y_adv5 = y.copy()

# Sample equal-sized benign + attack pools
benign_pool = X[y == 0].sample(n=20_000, random_state=42).reset_index(drop=True)
attack_pool = X[y == 1].sample(n=20_000, random_state=42).reset_index(drop=True)

# Blend ratio (Î± = benign weight)
alpha = 0.7
X_blended = alpha * benign_pool + (1 - alpha) * attack_pool
y_blended = pd.Series([1] * len(X_blended))  # still labeled as attack
```

```python
# Combine blended attacks with real benign flows
X_adv5_final = pd.concat([X[y == 0], X_blended])
y_adv5_final = pd.concat([y[y == 0], y_blended])

# Evaluate
y_pred = model.predict(X_adv5_final)
y_prob = model.predict_proba(X_adv5_final)[:, 1]
evaluate_attack(
    y_adv5_final, y_pred, y_prob,
    title="Grey-box Attack 5 â€“ Class Blending (70% Benign)"
)
```
